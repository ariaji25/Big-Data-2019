# Number Recognition

```python
from __future__ import absolute_import, division, print_function

#TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

The MNIST dataÂ¶

```python
#Download data 
from keras.datasets import mnist
(train_X,train_Y), (test_X,test_Y) = mnist.load_data('MNIST_data')
```

```python
print('Training data shape : ', train_X.shape, train_Y.shape)

print('Testing data shape : ', test_X.shape, test_Y.shape)
```

Show 25 Image of The MNIST data:

```python
class_names = ['Zero', 'One', 'Two', 'Three', 'Four', 
               'Five', 'Six', 'Seven', 'Eight', 'Nine']
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_X[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_Y[i]])
plt.show()
```

# ANN for Number Recognition

Data Preparation

```python
#Data normalization
train_X_t = train_X.reshape(-1, 28,28)
test_X_t = test_X.reshape(-1, 28,28)

train_X_t = train_X_t.astype('float32')
test_X_t = test_X_t.astype('float32')
train_X_t = train_X_t / 255.
test_X_t = test_X_t / 255.


print('Before Splitting:',train_X_t.shape, test_X_t.shape)

#Splitting data trainining
train_X_t,valid_X,train_label,valid_label = train_test_split(train_X_t, train_Y, test_size=0.3, random_state=13)
print('After Splitting:', train_X_t.shape,valid_X.shape,train_label.shape,valid_label.shape)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
batch_size=60
epochs=5
model.fit(train_X_t, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label))
```

```python
test_loss, test_acc = model.evaluate(test_X_t, test_Y)

print('Test accuracy:', test_acc)
```

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  
  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'
  
  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array[i], true_label[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1]) 
  predicted_label = np.argmax(predictions_array)
 
  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

Test Trainined model

```python
#Do Prediction
predictions = model.predict(test_X_t)

#Vew Results
num_rows = 10
num_cols = 6
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions, test_Y, test_X)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions, test_Y)
plt.show()
```

# CNN for Recognition

```python
#Data Preparation for CNN
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split

train_X_t = train_X.reshape(-1, 28,28, 1)
test_X_t = test_X.reshape(-1, 28,28, 1)

train_X_t = train_X_t.astype('float32')
test_X_t = test_X_t.astype('float32')
train_X_t = train_X_t / 255.
test_X_t = test_X_t / 255.

# Change the labels from categorical to one-hot encoding
train_Y_one_hot = to_categorical(train_Y)
test_Y_one_hot = to_categorical(test_Y)

# Display the change for category label using one-hot encoding
print('Original label:', train_Y[0])
print('After conversion to one-hot:', train_Y_one_hot[0])

print(train_X_t.shape, test_X_t.shape)
```

Create Model CNN:
Architecture of our network is:

(Input) -> [batch_size, 28, 28, 1] >> Apply 32 filter of [5x5]
(Convolutional layer 1) -> [batch_size, 28, 28, 32]
(ReLU 1) -> [?, 28, 28, 32]
(Max pooling 1) -> [?, 14, 14, 32]
(Convolutional layer 2) -> [?, 14, 14, 64]
(ReLU 2) -> [?, 14, 14, 64]
(Max pooling 2) -> [?, 7, 7, 64]
[fully connected layer 3] -> [1x1024]
[ReLU 3] -> [1x1024]
[Drop out] -> [1x1024]
[fully connected layer 4] -> [1x10]

```python
#Define CNN Model
import keras
from keras.models import Sequential,Input,Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU

digit_model = Sequential()
digit_model.add(Conv2D(32, kernel_size=(3, 3),activation='linear',input_shape=(28,28,1),padding='same'))
digit_model.add(LeakyReLU(alpha=0.1))
digit_model.add(MaxPooling2D((2, 2),padding='same'))
digit_model.add(Conv2D(64, (3, 3), activation='linear',padding='same'))
digit_model.add(LeakyReLU(alpha=0.1))
digit_model.add(MaxPooling2D(pool_size=(2, 2),padding='same'))
digit_model.add(Flatten())
digit_model.add(Dense(1024, activation='linear'))
digit_model.add(LeakyReLU(alpha=0.1))                  
digit_model.add(Dense(10, activation='softmax'))
```

```python
digit_model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adam(),metrics=['accuracy'])

digit_model.summary()
```

```python
train_X_t,valid_X,train_label,valid_label = train_test_split(train_X_t, train_Y_one_hot, test_size=0.3, random_state=13)
print(train_X_t.shape,valid_X.shape,train_label.shape,valid_label.shape)
```

```python
batch_size=60
epochs=2
digit_model.fit(train_X_t, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label))
```

```python
test_loss, test_acc = digit_model.evaluate(test_X_t, test_Y_one_hot)

print('Test accuracy:', test_acc)
```

```python
predictions = digit_model.predict(test_X_t)
# Plot the first X test images, their predicted label, and the true label
# Color correct predictions in blue, incorrect predictions in red
num_rows = 10
num_cols = 10

num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions, test_Y, test_X)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions, test_Y)
plt.show()
```

```python
idxTest=62
img=test_X[idxTest]
true_label=test_Y[idxTest]

img=img.reshape(-1,28,28, 1)
img = img.astype('float32')
img = img / 255.
predictions_single = digit_model.predict(img)
predicted_label = np.argmax(predictions_single)

plt.figure()
plt.imshow(test_X[idxTest], cmap=plt.cm.binary)
if predicted_label == true_label:
    color = 'blue'
else:
    color = 'red'  
plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_single),
                                class_names[true_label]),
                                color=color)
plt.show()
```
